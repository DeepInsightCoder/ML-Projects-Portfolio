# Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load the Dataset

```python
df = pd.read_csv('climate_change_indicators.csv')
```

# Display the First Few Rows

```python
df.head()
```

# Dataset Information

```python
# Show dataset information (data types, non-null counts, memory usage)
df.info()
```

```python
# Summary statistics for numerical columns (years data)
df.describe()
```

```python
# Check for missing values
df.isnull().sum()
```

#  Manipulating DataFrames

**Add a new column**

```python
# Calculate the mean for the yearly columns (assuming years are prefixed with 'F')
year_columns = df.columns[df.columns.str.startswith('F')]
df['Average_Indicator'] = df[year_columns].mean(axis=1)
```

**Filtering rows**

```python
# Filter countries with average indicator greater than 50
filtered_df = df[df['Average_Indicator'] > 50]
print(filtered_df[['Country', 'Average_Indicator']])
```

# Display the Cleaned Dataset

```python
print("\n=== Cleaned Dataset ===")
filtered_df.head()
```

**Handling missing values**

```python
# Fill missing values for the year columns using the mean
df[year_columns] = df[year_columns].fillna(df[year_columns].mean())
```

# Statistical Summary

```python
# Calculate and print mean, median, and standard deviation for a specific year's indicator (e.g., 2020)
print("Mean for 2020:", df['F2020'].mean())
print("Median for 2020:", df['F2020'].median())
print("Standard Deviation for 2020:", df['F2020'].std())
```

# Data Aggregation and Grouping

```python
# Group by Country and calculate the mean of the average indicator
grouped_data = df.groupby('Country')['Average_Indicator'].mean()
print(grouped_data)
```

# Data Visualization

**Histogram: To visualize the distribution of the indicator for 2020.**

```python
plt.figure(figsize=(10, 6))
df['F2020'].hist(bins=20)
plt.title('Distribution of Climate Indicator Values for 2020')
plt.xlabel('Indicator Value')
plt.ylabel('Frequency')
plt.grid(False)
plt.show()
```

**Scatter Plot: To visualize the relationship between two years (e.g., 2010 vs. 2020).**

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['F2010'], df['F2020'], alpha=0.5)
plt.title('Climate Indicator: 2010 vs 2020')
plt.xlabel('Indicator Value (2010)')
plt.ylabel('Indicator Value (2020)')
plt.grid(True)
plt.show()
```

**# Box Plot: To visualize the distribution of climate indicator values and identify outliers.**

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='F2020', data=df)
plt.title('Box Plot of Climate Indicator Values for 2020')
plt.xlabel('Indicator Value (2020)')
plt.grid(False)
plt.show()
```

**Pair Plot: To visualize the relationships between multiple numerical features.**

```python
# Selecting a few years to visualize relationships
selected_columns = ['F2010', 'F2015', 'F2020', 'Average_Indicator']
sns.pairplot(df[selected_columns])
plt.suptitle('Pair Plot of Climate Indicators', y=1.02)
plt.show()
```

**Heatmap: To show correlations between climate indicators over the years.**

```python
# Calculating correlations
correlation_matrix = df[selected_columns].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', square=True)
plt.title('Correlation Heatmap of Climate Indicators')
plt.show()
```

**Bar Plot: To visualize the average climate indicator for each country.**

```python
avg_indicators = df.groupby('Country')['Average_Indicator'].mean().sort_values(ascending=False)
plt.figure(figsize=(15, 10))
avg_indicators.head(10).plot(kind='bar', color='skyblue')
plt.title('Top 10 Countries by Average Climate Indicator')
plt.xlabel('Country')
plt.ylabel('Average Indicator Value')
plt.xticks(rotation=45)
plt.show()
```

**Facet Grid: To visualize the distribution of climate indicators for multiple countries.**

```python
# Selecting top 5 countries by average indicator
top_countries = df.groupby('Country')['Average_Indicator'].mean().nlargest(5).index
filtered_df = df[df['Country'].isin(top_countries)]

g = sns.FacetGrid(filtered_df, col='Country', col_wrap=3, height=4)
g.map(sns.histplot, 'F2020', kde=True)
g.add_legend()
plt.subplots_adjust(top=0.9)
g.fig.suptitle('Distribution of Climate Indicator Values for Top Countries in 2020')
plt.show()
```

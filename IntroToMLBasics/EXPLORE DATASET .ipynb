#  Manipulating DataFrames

**Add a new column**

```python
# Calculate the mean for the yearly columns (Feature Engineering)
year_columns = df.columns[df.columns.str.startswith('F')]
df['Average_Indicator'] = df[year_columns].mean(axis=1)
```

**Rename Columns**

```python
# Rename columns to authentic names
df.rename(columns={
    'ObjectId': 'ID', 
    'Country': 'Country_Name',
    'Indicator': 'Climate_Indicator',
    'Unit': 'Measurement_Unit',
    'Source': 'Data_Source'
}, inplace=True)
```

```python
# Rename year columns by removing the 'F' prefix
year_columns = [col for col in df.columns if col.startswith('F')]
df.rename(columns={col: col[1:] for col in year_columns}, inplace=True)
```

**Filter Column**

```python
# Filter countries with average indicator greater than 50
filtered_df = df[df['Average_Indicator'] > 1.0]
print(filtered_df[['Country_Name', 'Average_Indicator']])
```

**Reshape Data**

```python
# The year columns
year_columns = ['1961', '1962', '1963', '1964', '1965', '1966', '1967', '1968', '1969', '1970', 
                '1971', '1972', '1973', '1974', '1975', '1976', '1977', '1978', '1979', '1980', 
                '1981', '1982', '1983', '1984', '1985', '1986', '1987', '1988', '1989', '1990', 
                '1991', '1992', '1993', '1994', '1995', '1996', '1997', '1998', '1999', '2000', 
                '2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009', '2010', 
                '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', 
                '2021', '2022']
```

```python
df_melted = pd.melt(df, 
                    id_vars=['Country_Name', 'Climate_Indicator', 'Measurement_Unit', 'Data_Source'], 
                    value_vars=year_columns, 
                    var_name='Year', 
                    value_name='Indicator_Value')
```

```python
# Convert year column into numeric value
df_melted['Year'] = pd.to_numeric(df_melted['Year'])
```

**Remove Columns**

```python
columns_to_drop = ['ID', 'CTS_Code', 'CTS_Name', 'CTS_Full_Descriptor', 'Measurement_Unit' , 'Climate_Indicator',  'Climate_Indicator', 'Data_Source', 'ISO2', 'ISO3']
df.drop(columns=columns_to_drop, inplace=True)
```

**Handling missing values**

```python
# Fill missing values for the year columns using the mean
df_melted['Indicator_Value'].fillna(df_melted['Indicator_Value'].mean(), inplace=True)
```

**Display and Save the Cleaned Dataset**

```python
# Save the cleaned dataset
df.to_csv('Cleaned Climate Change Indicators.csv', index=False)
```

# Data Aggregation and Grouping

**Group BY**

```python
# Group by Country and calculate the mean of the average indicator
grouped_data = df.groupby('Country_Name')['Average_Indicator'].mean()
grouped_data
```

**Aggregate Yearly Data into Summary Features**

```python
# Calculate summary statistics
df['Mean_Indicator_Value'] = df.loc[:, '1961':'2022'].mean(axis=1)
df['Max_Indicator_Value'] = df.loc[:, '1961':'2022'].max(axis=1)
df['Min_Indicator_Value'] = df.loc[:, '1961':'2022'].min(axis=1)
```

# Data Visualization

```python
# visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns
```

**Distribution of the Climate Indicator Across All Countries:**
A histogram to visualize the distribution of the average climate indicator across all countries.

```python
plt.figure(figsize=(10,6))
sns.histplot(df['Average_Indicator'], kde=True, bins=30, color='skyblue')
plt.title('Distribution of Average Climate Indicator Across Countries')
plt.xlabel('Average Climate Indicator')
plt.ylabel('Frequency')
plt.show()
```

**Year-wise Indicator Trend for a Specific Country:**
A line plot to see how the climate indicator has changed over the years for a specific country.

```python
country_name = 'United States'
country_data = df_melted[df_melted['Country_Name'] == country_name]

plt.figure(figsize=(10, 6))
plt.plot(country_data['Year'], country_data['Indicator_Value'], marker='o')
plt.title(f'Climate Indicator Trend Over the Years for {country_name}')
plt.xlabel('Year')
plt.ylabel('Climate Indicator Value')
plt.xticks(rotation=45)
plt.show()
```

**Top 10 Countries with the Highest Average Indicato:r**
A pie chart to show the top 10 countries with the highest average climate indicator.

```python
top_countries = df.groupby('Country_Name')['Average_Indicator'].mean().nlargest(10)
top_countries.plot(kind='pie', autopct='%1.1f%%', figsize=(8, 8) , title=('Top 10 Countries with the Highest Average Climate Indicator') ,startangle=90)
plt.ylabel('')
plt.show()
```

**Correlation Heatmap:**
A heatmap to see the correlation between different years' climate indicators.

```python
plt.figure(figsize=(14,8))
sns.heatmap(df[year_columns].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Year-wise Climate Indicators')
plt.show()
```

**Country-wise Mean Indicator for Multiple Countries:**
A multi-line plot to compare the climate indicators over time for multiple countries.

```python
countries = ['United States', 'China', 'India', 'Brazil', 'Russia']
plt.figure(figsize=(14,8))

for country in countries:
    country_data = df_melted[df_melted['Country_Name'] == country]
    sns.lineplot(x='Year', y='Indicator_Value', data=country_data, label=country)

plt.title('Climate Indicator Trend for Multiple Countries')
plt.xlabel('Year')
plt.ylabel('Climate Indicator Value')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

**Boxplot of Climate Indicator Across Countries:**
A boxplot to show the distribution of climate indicator values and identify outliers.

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='2020', data=df)
plt.title('Box Plot of Climate Indicator Values for 2020')
plt.xlabel('Indicator Value (2020)')
plt.grid(False)
plt.show()
```

**Scatter Plot:** To visualize the relationship between two years (2010 vs. 2020).

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['2010'], df['2020'], alpha=0.5)
plt.title('Climate Indicator: 2010 vs 2020')
plt.xlabel('Indicator Value (2010)')
plt.ylabel('Indicator Value (2020)')
plt.grid(True)
plt.show()
```

**Pair Plot:** To visualize the relationships between multiple numerical features.

```python
selected_columns = ['2010', '2015', '2020', 'Average_Indicator']
sns.pairplot(df[selected_columns])
plt.suptitle('Pair Plot of Climate Indicators', y=1.02)
plt.show()
```

**Heatmap:** To show correlations between climate indicators over the years(selected years).

```python
correlation_matrix = df[selected_columns].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', square=True)
plt.title('Correlation Heatmap of Climate Indicators')
plt.show()
```
